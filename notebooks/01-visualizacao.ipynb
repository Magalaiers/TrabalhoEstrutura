```python
# Importa a biblioteca que instalamos
import pandas as pd
import os

# ---- CONFIGURE AQUI ----
NOME_DO_ARQUIVO_DO_DATASET = "heart_attack_prediction_dataset.csv" # <--- SUBSTITUA PELO NOME REAL DO SEU ARQUIVO
# --------------------------

# Constrói o caminho para o arquivo de forma robusta
# Isso garante que funcione independentemente de onde você execute o notebook
caminho_do_arquivo = os.path.join("..", "dataset", NOME_DO_ARQUIVO_DO_DATASET)

# Tenta carregar o dataset
try:
    df = pd.read_csv(caminho_do_arquivo)

    print("✅ Dataset carregado com sucesso!")
    print(f"O dataset tem {df.shape[0]} linhas (amostras) e {df.shape[1]} colunas (features).")

    print("\n--- Primeiras 5 linhas do dataset: ---")
    display(df.head()) # 'display' é melhor que 'print' para dataframes em notebooks

    print("\n--- Informações sobre as colunas e tipos de dados: ---")
    df.info()

except FileNotFoundError:
    print(f"❌ ERRO: Arquivo não encontrado!")
    print(f"Verifique se o nome '{NOME_DO_ARQUIVO_DO_DATASET}' está correto e se o arquivo está na pasta 'dataset/'.")
except Exception as e:
    print(f"❌ ERRO ao carregar o dataset: {e}")
```

```python
# --- Parte 2: Estatísticas Descritivas Gerais ---
print("\n" + "="*50)
print("   ESTATÍSTICAS DESCRITIVAS (COLUNAS NUMÉRICAS)")
print("="*50)
# O .T (transpor) facilita a leitura da tabela
display(df.describe().T)
```

```python
# --- Parte 3: Análise de Frequência de Colunas Categóricas ---
print("\n" + "="*50)
print("   DISTRIBUIÇÃO DE DADOS CATEGÓRICOS")
print("="*50)

# Lista de colunas categóricas de interesse
colunas_categoricas = ['Sex', 'Diet', 'Diabetes', 'Smoking', 'Heart Attack Risk']

for coluna in colunas_categoricas:
    if coluna in df.columns:
        print(f"\n--- Frequência para a coluna: '{coluna}' ---")
        display(df[coluna].value_counts())
```

```python
# --- Parte 1: Pré-processamento da Coluna 'Blood Pressure' ---
print("\nProcessando a coluna 'Blood Pressure'...")
if 'Blood Pressure' in df.columns:
    # Separa a string "sistólica/diastólica" em duas novas colunas
    split_bp = df['Blood Pressure'].str.split('/', expand=True)
    df['Pressao_Sistolica'] = pd.to_numeric(split_bp[0], errors='coerce')
    df['Pressao_Diastolica'] = pd.to_numeric(split_bp[1], errors='coerce')
    
    # Remove a coluna original de texto
    df = df.drop(columns=['Blood Pressure'])
    print("✅ Coluna 'Blood Pressure' transformada em 'Pressao_Sistolica' e 'Pressao_Diastolica'.")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# --- Parte 4: Visualização da Distribuição de Features Numéricas ---
print("\n" + "="*50)
print("   HISTOGRAMAS DE DISTRIBUIÇÃO (FEATURES NUMÉRICAS)")
print("="*50)

# Lista de colunas numéricas de interesse para visualizar
colunas_numericas_vis = ['Age', 'Cholesterol', 'Heart Rate', 'BMI']

# Cria uma figura com múltiplos subplots
fig, axes = plt.subplots(2, 2, figsize=(16, 10))
fig.suptitle('Distribuição de Features Numéricas Chave', fontsize=18)
# Transforma os eixos 2x2 em uma lista 1D para facilitar o loop
axes = axes.flatten()

for i, col in enumerate(colunas_numericas_vis):
    sns.histplot(df[col], kde=True, ax=axes[i], bins=30)
    axes[i].set_title(f'Distribuição de {col}')

plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()
```

```python
# ==============================================================================
# PARTE 2: VISUALIZAÇÃO DE VARIÁVEIS NUMÉRICAS (Apenas Histogramas)
# ==============================================================================
print("\nGerando dashboard para variáveis numéricas...")

df_numericas = df.select_dtypes(include=np.number)

# Define um layout de grade mais compacto
n_cols = 3
n_rows = (len(df_numericas.columns) + n_cols - 1) // n_cols
fig, axes = plt.subplots(n_rows, n_cols, figsize=(18, n_rows * 4))
fig.suptitle('Análise de Distribuição das Variáveis Numéricas', fontsize=20, y=1)
axes = axes.flatten() # Transforma a grade 2D em uma lista 1D para facilitar o loop

for i, col_name in enumerate(df_numericas.columns):
    ax = axes[i]
    sns.histplot(df_numericas[col_name], kde=True, ax=ax, bins=30)
    ax.set_title(f'Distribuição de {col_name}')

# Esconde os eixos que não foram usados
if len(df_numericas.columns) < len(axes):
    for j in range(len(df_numericas.columns), len(axes)):
        axes[j].set_visible(False)
            
plt.tight_layout(rect=[0, 0, 1, 0.98])
plt.show()
```

```python
# ==============================================================================
# PARTE 3: VISUALIZAÇÃO DE VARIÁVEIS CATEGÓRICAS
# ==============================================================================
print("\nGerando dashboard para variáveis categóricas...")

df_categoricas = df.select_dtypes(include=['object']).drop(columns=['Patient ID'], errors='ignore')

if not df_categoricas.empty:
    n_cols_cat = 2
    n_rows_cat = (len(df_categoricas.columns) + n_cols_cat - 1) // n_cols_cat
    fig_cat, axes_cat = plt.subplots(n_rows_cat, n_cols_cat, figsize=(16, n_rows_cat * 5))
    fig_cat.suptitle('Análise de Frequência das Variáveis Categóricas', fontsize=20, y=1.02)
    axes_cat = axes_cat.flatten()

    for i, col_name in enumerate(df_categoricas.columns):
        ax = axes_cat[i]
        sns.countplot(y=df_categoricas[col_name], ax=ax, order = df_categoricas[col_name].value_counts().index, palette='viridis')
        ax.set_title(f'Contagem de {col_name}')
        ax.set_xlabel('Contagem')
        ax.set_ylabel('')
        for container in ax.containers:
            ax.bar_label(container)

    if len(df_categoricas.columns) < len(axes_cat):
        for j in range(len(df_categoricas.columns), len(axes_cat)):
            axes_cat[j].set_visible(False)
            
    plt.tight_layout(rect=[0, 0, 1, 0.98])
    plt.show()
else:
    print("Nenhuma outra coluna categórica para plotar.")
```

```python
# --- Parte 5: Matriz de Correlação ---
print("\n" + "="*50)
print("   MATRIZ DE CORRELAÇÃO ENTRE FEATURES NUMÉRICAS")
print("="*50)

# Calcula a matriz de correlação
corr_matrix = df.corr(numeric_only=True)

# Cria o mapa de calor para visualizar a matriz
plt.figure(figsize=(16, 12))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f") # annot=False para não poluir com números
plt.title('Mapa de Calor da Correlação entre Variáveis', fontsize=18)
plt.show()
```
